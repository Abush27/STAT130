````python
import numpy as np

# Simulate ages of 150 attendees at a music festival
np.random.seed(42)  # For reproducibility
# The majority of attendees are young adults
young_adults = np.random.normal(loc=25, scale=3, size=120)
# A few younger teenagers
teenagers = np.random.normal(loc=17, scale=2, size=20)
# A small number of older attendees
older_attendees = np.random.normal(loc=50, scale=5, size=10)
# Combine all groups into one dataset
festival_ages = np.concatenate([young_adults, teenagers, older_attendees])

data = festival_ages

# Number of bootstrap samples
n_bootstraps = 10000

# Function to generate bootstrapped sample statistics (mean by default)
def bootstrap_statistic(data, n_bootstraps, stat_func=np.mean):
    boot_stats = []
    n = len(data)
    for _ in range(n_bootstraps):
        # Resample with replacement from the original data
        boot_sample = np.random.choice(data, size=n, replace=True)
        # Calculate the statistic (mean by default, can change to median, std, etc.)
        boot_stats.append(stat_func(boot_sample))
    return np.array(boot_stats)

# Generate bootstrapped means (change np.mean to np.median for median)
boot_means = bootstrap_statistic(data, n_bootstraps, stat_func=np.mean)

# Calculate the 2.5th and 97.5th percentiles for the 95% confidence interval
lower_bound = np.percentile(boot_means, 2.5)
upper_bound = np.percentile(boot_means, 97.5)

# Output the confidence interval for the mean
print(f"95% bootstrapped confidence interval for the mean: [{lower_bound}, {upper_bound}]")
````

````python
Neverthless, these inferences made from our sample are not completely represntive of the entire population.
The graph earlier showed us the scoreChange column with a mean of 3.3. This means from our sample, the average person's health 
score will increase by 3.3 once they take the vaccine. However since our sample does not represent the whole population we
cannot assume its accuracy right away. Hence we can create boostrapped samples now to get a histogram of means for further 
calculations of the confidence intervals. 
````

````python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import norm

# Calculate Score Change
df['ScoreChange'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Bootstrapping process
n_bootstraps = 1000  # Number of bootstrap samples
bootstrap_means = []

for _ in range(n_bootstraps):
    # Sample with replacement
    sample = df['ScoreChange'].sample(n=len(df), replace=True)
    bootstrap_means.append(sample.mean())

# Convert bootstrap means to a numpy array for easier handling
bootstrap_means = np.array(bootstrap_means)

# Create a histogram of the bootstrapped means
fig = px.histogram(
    x=bootstrap_means,
    histnorm='probability density',
    nbins=30,
    title='Bootstrapped Means of Score Change',
    labels={'x': 'Bootstrapped Means', 'y': 'Probability Density'},
    opacity=0.6,
    color_discrete_sequence=['orange']
)

# Add a vertical line for the bootstrapped mean
bootstrap_mean = np.mean(bootstrap_means)
fig.add_vline(x=bootstrap_mean, line_dash="dot", annotation_text=f'Mean: {bootstrap_mean:.2f}', 
              line_color='red')

# Add lines for the 95% confidence interval of the bootstrapped means
bootstrap_ci_lower = np.percentile(bootstrap_means, 2.5)
bootstrap_ci_upper = np.percentile(bootstrap_means, 97.5)

fig.add_vline(x=bootstrap_ci_lower, line_dash="dash", annotation_text='2.5% CI', 
              line_color='red')
fig.add_vline(x=bootstrap_ci_upper, line_dash="dash", annotation_text='97.5% CI', 
              line_color='red')

# Add a vertical line at 0 for reference
fig.add_vline(x=0, line_dash="solid", annotation_text='', 
              line_color='green')

# Fit a normal distribution to the bootstrapped means
mu, sigma = norm.fit(bootstrap_means)

# Generate the x values (support) for the PDF
x_values = np.linspace(min(bootstrap_means), max(bootstrap_means), 500)

# Calculate the y values for the PDF
pdf_values = norm.pdf(x_values, mu, sigma)

# Add the PDF curve to the histogram
fig.add_trace(go.Scatter(x=x_values, y=pdf_values, mode='lines', 
                         name='Fitted Normal PDF', line_color='blue'))

# Convert the figure to a static image
img_bytes = pio.to_image(fig, format='png')  # Convert to PNG bytes

# Display the static image
display(Image(img_bytes))
````

````python
The graph above demonstrates the histogram of the means with a confidence interval created by adding the red dotted lines
at the 2.5 and 97.5 percentile, in order to cover 95% of the data around the mean. The graph shape seems to be a normal 
distribution, as most of the data is is clustered at the mean and almost evenely layered around it. This bell curve demonstrates
that the closer the value is to the mean, the more likely it is to appear, meaning it is represented quite acurately. 

I also created a boxplot below to better visualise how the data is clustered around the quartiles, especially the confidence 
intervals created at 2.5 and 97.5 percentiles.
````
